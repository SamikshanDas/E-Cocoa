## Univariate Selection

Statistical tests can be used to select those features that have the strongest relationship with the output variable.

The scikit-learn library provides the SelectKBest class that can be used with a suite of different statistical tests to select a specific number of features.

The example below uses the chi-squared (chi²) statistical test for non-negative features to select 15 of the best features.

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2,mutual_info_classif
```

```python
#apply SelectKBest class to extract top 16 best features
bestfeatures = SelectKBest(score_func=mutual_info_classif,k=23)
fit = bestfeatures.fit(x_norm,y)
```

```python
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(x_norm.columns)
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['features','Score']  #naming the dataframe columns
```

```python
featureScores
```

```python
fl=featureScores.nlargest(23,'Score')
fl
```

```python
print(featureScores.nlargest(23,'Score'))  #print 16 best features top to bottom
```

```python
feat_score = pd.Series(fit.scores_, index=x_norm.columns)
plt.subplots(figsize=(10,5))
feat_score.nlargest(23).plot(kind='barh')
plt.xlabel('Scores')
plt.ylabel('features')
plt.show()
```

# Feature Importance
You can get the feature importance of each feature of your dataset by using the feature importance property of the model.

Feature importance gives you a score for each feature of your data, the higher the score more important or relevant is the feature towards your output variable.

Feature importance is an inbuilt class that comes with Tree Based Classifiers, we will be using Extra Tree Classifier for extracting the top 15 features for the dataset.

```python
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model = ExtraTreesClassifier()
model.fit(x_norm,y)
```

```python
print(model.feature_importances_) #use inbuilt class feature_importances of tree based classifiers
```

```python
fscore=pd.DataFrame(model.feature_importances_)
fcolumn=pd.DataFrame(x_norm.columns)
fimportance=pd.concat([fcolumn,fscore],axis=1)
fimportance.columns=['Features','Importance']
fimportance
```

```python
#plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=x_norm.columns)
plt.subplots(figsize=(10,5))
feat_importances.nlargest(23).plot(kind='barh')
plt.xlabel('Importance')
plt.ylabel('features')
plt.show()
```

```python
fl1=feat_importances.nlargest(23)
plt.subplots(figsize=(10,5))
plt.plot([i for i in fl1],[j for j in fl1.index],'b')
plt.xlabel('Importance')
plt.ylabel('features')
plt.show()
```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
```

```python
def getscore(model,x,y):
    skfold=StratifiedKFold(n_splits=10)
    scores=cross_val_score(model,x,y,cv=skfold)
    print("Scores:",scores)
    print("Max:","{:.2}".format(np.max(scores)))
    print("Min:","{:.2}".format(np.min(scores)))
    print("Mean:","{:.2}".format(np.mean(scores)))
```

```python
print("Knn CRoss validation:")
getscore(neigh,x_data,y_data)
```

```python
print("Decision Tree CRoss validation score:",end=" ")
getscore(ctree,X,Y)
```

```python
print("SVM CRoss validation score:",end=" ")
getscore(clf,x_data,y_data)
```

```python
print("Random Forest CRoss validation score:",end=" ")
getscore(rf,X,Y)
```
